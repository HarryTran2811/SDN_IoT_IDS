```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '35000_each.csv'
df = pd.read_csv(file_path)
df.head()
```

Drop column not needed and Handle Missing Values

```python
# Check for missing values
missing_values = df.isnull().sum()
print(f"Missing values per column:\n{missing_values[missing_values > 0]}")
```

Check whether srcPort and dstPort are skewed or not

```python
# Plot the distribution of srcPort and dstPort columns
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['srcPort'].dropna(), kde=True)
plt.title('Distribution of srcPort')

plt.subplot(1, 2, 2)
sns.histplot(df['dstPort'].dropna(), kde=True)
plt.title('Distribution of dstPort')

plt.tight_layout()
plt.show()
```

```python
data_clean = df.drop(columns=['srcMAC', 'dstMAC', 'srcIP', 'dstIP'])
```

```python
data_clean['srcPort'].fillna(data_clean['srcPort'].median(), inplace=True)
data_clean['dstPort'].fillna(data_clean['dstPort'].median(), inplace=True)
```

Convert Data Types

```python
data_clean['srcPort'] = data_clean['srcPort'].astype(int)
data_clean['dstPort'] = data_clean['dstPort'].astype(int)
data_clean['Dur'] = data_clean['Dur'].astype(float)
data_clean['Protocol'] = data_clean['Protocol'].astype('category')
data_clean['Attack'] = data_clean['Attack'].astype('category')
data_clean['Category'] = data_clean['Category'].astype('category')
```

```python
# Check for duplicate records
duplicates = data_clean.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")
```

```python
data_clean.drop_duplicates(inplace=True)
```

```python
duplicates_after = data_clean.duplicated().sum()
print(f"Number of duplicate rows after cleaning: {duplicates_after}")
```

```python
import numpy as np
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import TruncatedSVD
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout
from tensorflow.keras.optimizers import Adam
from scipy.sparse import csr_matrix
from sklearn.model_selection import train_test_split

# Assuming X_train_encoded and X_test are initially sparse matrices
# Make sure these are CSR or CSC sparse matrices for TruncatedSVD to work correctly

# Step 1: Ensure both X_train_encoded and X_test are in the CSR sparse format
if not isinstance(X_train_encoded, csr_matrix):
    X_train_encoded = csr_matrix(X_train_encoded)
if not isinstance(X_test, csr_matrix):
    X_test = csr_matrix(X_test)

# Step 2: Apply TruncatedSVD to reduce dimensionality
svd = TruncatedSVD(n_components=100)  # Adjust n_components based on memory and performance needs
X_train_reduced = svd.fit_transform(X_train_encoded)
X_test_reduced = svd.transform(X_test)

# Step 3: Use RandomOverSampler to balance the dataset with less memory usage
ros = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = ros.fit_resample(X_train_reduced, y_train)

# Step 4: Convert to DataFrame for consistency if needed
# Directly converting to DataFrame might not be necessary; proceed with NumPy arrays
X_train_resampled = pd.DataFrame(X_train_resampled)
X_test_reduced = pd.DataFrame(X_test_reduced)

# Step 5: Standardize the data
scaler = StandardScaler()
X_train_resampled_scaled = scaler.fit_transform(X_train_resampled)
X_test_scaled = scaler.transform(X_test_reduced)

# Step 6: Reshape data for LSTM input (samples, timesteps, features)
X_train_resampled_lstm = np.reshape(X_train_resampled_scaled, (X_train_resampled_scaled.shape[0], 1, X_train_resampled_scaled.shape[1]))
X_test_lstm = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# Step 7: Define the BiLSTM model
model = Sequential()
model.add(Bidirectional(LSTM(64, return_sequences=True, activation='relu'), input_shape=(1, X_train_resampled_lstm.shape[2])))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(32, activation='relu')))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))

# Step 8: Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Step 9: Train the BiLSTM model
history = model.fit(X_train_resampled_lstm, y_train_resampled, epochs=50, batch_size=64, validation_split=0.2, verbose=1)

# Step 10: Evaluate the model
loss, accuracy = model.evaluate(X_test_lstm, y_test)
print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")
```
